```python
#Métricas 
import pandas as pd
import altair as alt
import numpy as np

# Lectura del fichero JSON en el respositorio abierto de QuijotESTILO en GitHub con los datos de métricas
fichero = 'https://raw.githubusercontent.com/st-srnm/quijotestilo/refs/heads/main/QUIJOTESTILO-Quijote-metricas-ALTAIR-03.json'
metricas_df = pd.read_json(fichero)    #esto es para leer el fichero json

#Convertir valores numéricos con como a numércios con punto versión nueva ya pasa con "."metricas_df['Densidad_lexica'] = (
metricas_df['Longitud_media_frases'].astype(str).str.replace(',', '.', regex=False).astype(float)
#metricas_df['Densidad_lexica'] = str(metricas_df['Densidad_lexica']).replace(',','.').float()

metricas_df['Indice_Legilibilidad_FH'].astype(str).str.replace(',', '.', regex=False).astype(float)
metricas_df['Densidad_lexica'] = (
    metricas_df['Densidad_lexica']
    .astype(str)
    .str.replace(',', '.', regex=False)
    .astype(float)
)
print(metricas_df['Densidad_lexica'])



# Asignar título y subtítulo al gráfico
titulo = ['Densidad léxica por capítulo del Quijote', 'QuijotESTILO']

#Extaer Parte del Capitulo para alt.Color
##metricas_df['Parte'] = metricas_df['Capitulo'].str[0:1]
#metricas_df['Parte1-2'] = np.where(metricas_df['Parte'].astype(str).str.startswith("1"), 'Primera Parte', 'Segunda Parte')
metricas_df['Capitulo_num'] = metricas_df['Capitulo'].str[0:5]
metricas_df['media_dens_lex'] = str(round(metricas_df['Total_palabras'].mean(),1))


#Ordeción metricas_df por GrupoQS0-QSx y luego por Capitulo_num
metricas_df_ordenado = metricas_df.sort_values(['GrupoQS0-QSx', 'Capitulo_num'])
metricas_df_ordenado['aux_y_orden'] = metricas_df['GrupoQS0-QSx'] + metricas_df['Capitulo_num']

# Creación del gráfico de barras de ALTAIR con sus parámetros
metricas = alt.Chart(metricas_df_ordenado).mark_bar().encode(
    alt.X('Densidad_lexica:Q', title = 'Densidad léxica por capítulo grupos Q00S00 y QnnSnn', axis=alt.Axis(labelFontSize=12)),
    alt.Y('Capitulo_num:O', title='Densidad léxica por capítulo grupos Q00S00 y QnnSnn', sort=alt.EncodingSortField(field='aux_y_orden', order='ascending')),
          color=alt.Color('GrupoQS0-QSx:N', title='Q00S00 y QnnSnn'),
          tooltip= ['Parte', 'Capitulo_num', 'Densidad_lexica', 'Total_palabras', 'media_dens_lex', 'GrupoQS0-QSx', 'Frecuencia_Quijote', 'Frecuencia_Sancho']   # Información que se visualiza al poner el ratón encima del gráfico
).properties(
    title = titulo,
    width = 600,
    height= 1200
    )
#Gráfica con linea de puntos con media densidad léxica por capitulo
aux_media = str(round(metricas_df['Densidad_lexica'].mean(),2))
puntos_media_df = pd.DataFrame({
    'x': [aux_media] * len(metricas_df['Densidad_lexica']),
    'Capitulo_num': metricas_df['Capitulo_num']
})

puntos_media = alt.Chart(puntos_media_df).mark_point(filled=True, color='red', size=12).encode(
    x=alt.X('x:Q'),
    y=alt.Y('Capitulo_num:O')
)
#Texto con el número del valor medio en la parte superior
texto_df = pd.DataFrame({
    'x': [aux_media],
    'y': [metricas_df['Capitulo_num'].iloc[0]]  # Toma el primer valor del eje y solo para ubicar el texto
})
texto = alt.Chart(texto_df).mark_text(
    text=str('Media = ' + aux_media),
    fontSize=12,
    fontWeight='bold',
    color='red',
    dy=-3  # Ajusta este valor para mover el texto más arriba
).encode(
    x=alt.X('x:Q'),
    y=alt.value(0)  # Coloca el texto en la parte superior del gráfico
)
#Inicialización variable texto por capítulos
aux_densidad_lexica = metricas_df.groupby('GrupoQS0-QSx')['Densidad_lexica'].mean()
aux_densidad_lexica1 = round(aux_densidad_lexica.loc['Q00S00'],2)   # si 'Parte' es numérico
aux_densidad_lexica2 = round(aux_densidad_lexica.loc['QnnSnn'],2)
#aux_densidad_lexica1 = aux_densidad_lexica['Densidadlexica'].iloc[0]
#aux_densidad_lexica2 = aux_densidad_lexica['Densidadlexica'].iloc[1]
texto_parte1 = pd.DataFrame({'t':['Q00S00. Media Densidad léxica: ' + str(aux_densidad_lexica1)]})
#Texto media densidad léxica parte 1
parte1 = alt.Chart(texto_parte1).mark_text(
    baseline='top',
    fontSize = 14,
    fontWeight='bold',
    color = 'white',
).encode(
    x=alt.value(250),
    y=alt.value(30),
    text = 't'
)
#Texto total palabras parte 2
texto_parte2 = pd.DataFrame({'t':['QnnSnn. Media Densidad léxica: ' + str(aux_densidad_lexica2)]})
parte2 = alt.Chart(texto_parte2).mark_text(
    baseline='top',
    fontSize = 14,
    fontWeight='bold',
    color = 'white',
).encode(
    x=alt.value(250),
    y=alt.value(100),
    text = 't'
)

#Visualizar los gráficos
#grafico_completo = metricas + puntos_media + texto +parte1 + parte2
#grafico_completo.display()
grafico_parte = metricas + texto + parte1 + parte2
grafico_parte.display()
```
