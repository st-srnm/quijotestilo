```python
import re
import requests
from io import StringIO
import os

def roman_to_int(roman):
    roman_values = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000}
    total = 0
    prev_value = 0
    for char in reversed(roman):
        current_value = roman_values[char]
        if current_value >= prev_value:
            total += current_value
        else:
            total -= current_value
        prev_value = current_value
    return total

def dividir_quijote_en_capitulos(url):
    # Descargar el contenido del archivo
    response = requests.get(url)
    texto_completo = response.text

 # Dividir el texto en partes
    partes = re.split(r'(Primera parte del ingenioso hidalgo don Quijote de la Mancha|Segunda parte del ingenioso caballero don Quijote de la Mancha)', texto_completo)
    parte_actual = 0  # Variable para rastrear la parte actual (1 o 2)

    for parte in partes[1:]:
        if parte.startswith("Primera parte"):
            parte_actual = 1
            continue
        elif parte.startswith("Segunda parte"):
            parte_actual = 2
            continue

        if parte_actual == 0:  # Saltar si no estamos en una parte válida
            continue

        # Encontrar todos los capítulos
        capitulos = re.split(r'(Capítulo [IVXLCDM]+\..*)', parte)
        for i in range(1, len(capitulos), 2):
            titulo = capitulos[i]
            contenido = capitulos[i+1] if i+1 < len(capitulos) else ""

            # Extraer el número de capítulo en romano
            match = re.search(r'Capítulo ([IVXLCDM]+)', titulo)
            if match:
                num_capitulo = roman_to_int(match.group(1))
                # Crear el nombre del archivo
                nombre_archivo = f"{parte_actual}-C{num_capitulo:02d}.txt"
                carpeta = os.path.join('Quijote', 'Quijote por capitulos')
                ruta_completa = os.path.join(carpeta, nombre_archivo)
                print('nombre_archivo = ', nombre_archivo)
                # Escribir el contenido en el archivo
                with open(ruta_completa, 'w', encoding='utf-8') as f:
                    f.write(titulo + "\n\n" + contenido.strip())

                print(f"Creado: {nombre_archivo}")

# URL del archivo de texto del Quijote
url_raiz = "https://raw.githubusercontent.com/st-srnm/quijotestilo/refs/heads/main/"
github_fichero ='Quijote-TextoCompleto-Parte1y2-SinIntroducciones-01.txt'
url = url_raiz + github_fichero
os.makedirs('/QuijotESTILO/Quijote/Quijote por capitulos/', exist_ok=True)
carpeta = os.path.join('Quijote', 'Quijote por capitulos')



dividir_quijote_en_capitulos(url)
```
