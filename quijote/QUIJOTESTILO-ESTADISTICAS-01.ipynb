```python
import pandas as pd
from scipy import stats
import numpy as np

#leer fichero
fichero = 'https://raw.githubusercontent.com/st-srnm/quijotestilo/refs/heads/main/QUIJOTESTILO-Quijote-metricas-ALTAIR-03.json'
metricas_df = pd.read_json(fichero)    #esto es para leer el fichero json
#Convertir valores numéricos con como a numércios con punto
def safe_to_numeric(col):
    if col.dtype == 'object':
        try:
            return pd.to_numeric(col.str.replace(',', '.', regex=False))
        except Exception:
            return col
    else:
        return col

#Cambiar los "," por "." en varia les con decimales
metricas_df = metricas_df.apply(safe_to_numeric)

#Análisis estadísico
#
#Total_caracteres
print('\n******************************************')
variable = 'Total_caracteres'

valores_df = metricas_df
# T-test para dos grupos
grupos = valores_df['GrupoQS0-QSx'].unique()
grupo1 = valores_df[valores_df['GrupoQS0-QSx'] == grupos[0]][variable]
grupo2 = valores_df[valores_df['GrupoQS0-QSx'] == grupos[1]][variable]
t_stat, p_value = stats.ttest_ind(grupo1, grupo2)
print('Variable = ', variable, '\n')
print(f"Se utilizó el t-test de Student para muestras independientes.")
print(f"El estadístico t es {t_stat:.3f} y el valor p es {p_value:.4f}.")
if p_value < 0.05:
     print("Existe diferencia estadísticamente significativa entre los grupos de B.")
else:
    print("No se encontró diferencia estadísticamente significativa entre los grupos de B.")
print('\n******************************************')

#
#
#Total_palabras
variable = 'Total_palabras'

valores_df = metricas_df
# T-test para dos grupos
grupos = valores_df['GrupoQS0-QSx'].unique()
grupo1 = valores_df[valores_df['GrupoQS0-QSx'] == grupos[0]][variable]
grupo2 = valores_df[valores_df['GrupoQS0-QSx'] == grupos[1]][variable]
t_stat, p_value = stats.ttest_ind(grupo1, grupo2)
print('Variable = ', variable, '\n')
print(f"Se utilizó el t-test de Student para muestras independientes.")
print(f"El estadístico t es {t_stat:.3f} y el valor p es {p_value:.4f}.")
if p_value < 0.05:
     print("Existe diferencia estadísticamente significativa entre los grupos de B.")
else:
    print("No se encontró diferencia estadísticamente significativa entre los grupos de B.")
print('\n******************************************')

#
#
#Longitud_media_frases
variable = 'Longitud_media_frases'

valores_df = metricas_df
# T-test para dos grupos
grupos = valores_df['GrupoQS0-QSx'].unique()
grupo1 = valores_df[valores_df['GrupoQS0-QSx'] == grupos[0]][variable]
grupo2 = valores_df[valores_df['GrupoQS0-QSx'] == grupos[1]][variable]
t_stat, p_value = stats.ttest_ind(grupo1, grupo2)
print('Variable = ', variable, '\n')
print(f"Se utilizó el t-test de Student para muestras independientes.")
print(f"El estadístico t es {t_stat:.3f} y el valor p es {p_value:.4f}.")
if p_value < 0.05:
     print("Existe diferencia estadísticamente significativa entre los grupos de B.")
else:
    print("No se encontró diferencia estadísticamente significativa entre los grupos de B.")
print('\n******************************************')

#
#
#Palabras_lexicas
variable = 'Palabras_lexicas'

valores_df = metricas_df
# T-test para dos grupos
grupos = valores_df['GrupoQS0-QSx'].unique()
grupo1 = valores_df[valores_df['GrupoQS0-QSx'] == grupos[0]][variable]
grupo2 = valores_df[valores_df['GrupoQS0-QSx'] == grupos[1]][variable]
t_stat, p_value = stats.ttest_ind(grupo1, grupo2)
print('Variable = ', variable, '\n')
print(f"Se utilizó el t-test de Student para muestras independientes.")
print(f"El estadístico t es {t_stat:.3f} y el valor p es {p_value:.4f}.")
if p_value < 0.05:
     print("Existe diferencia estadísticamente significativa entre los grupos de B.")
else:
    print("No se encontró diferencia estadísticamente significativa entre los grupos de B.")
print('\n******************************************')

#
#
#Palabras_funcionales
variable = 'Palabras_funcionales'

valores_df = metricas_df
# T-test para dos grupos
grupos = valores_df['GrupoQS0-QSx'].unique()
grupo1 = valores_df[valores_df['GrupoQS0-QSx'] == grupos[0]][variable]
grupo2 = valores_df[valores_df['GrupoQS0-QSx'] == grupos[1]][variable]
t_stat, p_value = stats.ttest_ind(grupo1, grupo2)
print('Variable = ', variable, '\n')
print(f"Se utilizó el t-test de Student para muestras independientes.")
print(f"El estadístico t es {t_stat:.3f} y el valor p es {p_value:.4f}.")
if p_value < 0.05:
     print("Existe diferencia estadísticamente significativa entre los grupos de B.")
else:
    print("No se encontró diferencia estadísticamente significativa entre los grupos de B.")
print('\n******************************************')

#
#
#Densidad léxica
variable = 'Densidad_lexica'

valores_df = metricas_df
# T-test para dos grupos
grupos = valores_df['GrupoQS0-QSx'].unique()
grupo1 = valores_df[valores_df['GrupoQS0-QSx'] == grupos[0]][variable]
grupo2 = valores_df[valores_df['GrupoQS0-QSx'] == grupos[1]][variable]
t_stat, p_value = stats.ttest_ind(grupo1, grupo2)
print('Variable = ', variable, '\n')
print(f"Se utilizó el t-test de Student para muestras independientes.")
print(f"El estadístico t es {t_stat:.3f} y el valor p es {p_value:.4f}.")
if p_value < 0.05:
     print("Existe diferencia estadísticamente significativa entre los grupos de B.")
else:
    print("No se encontró diferencia estadísticamente significativa entre los grupos de B.")
print('\n******************************************')


#
#
#Indice_Legilibilidad_FH
variable = 'Indice_Legilibilidad_FH'

valores_df = metricas_df
# T-test para dos grupos
grupos = valores_df['GrupoQS0-QSx'].unique()
grupo1 = valores_df[valores_df['GrupoQS0-QSx'] == grupos[0]][variable]
grupo2 = valores_df[valores_df['GrupoQS0-QSx'] == grupos[1]][variable]
t_stat, p_value = stats.ttest_ind(grupo1, grupo2)
print('Variable = ', variable, '\n')
print(f"Se utilizó el t-test de Student para muestras independientes.")
print(f"El estadístico t es {t_stat:.3f} y el valor p es {p_value:.4f}.")
if p_value < 0.05:
     print("Existe diferencia estadísticamente significativa entre los grupos de B.")
else:
    print("No se encontró diferencia estadísticamente significativa entre los grupos de B.")
print('\n******************************************')

#
#
#Frecuencia_Quijote
variable = 'Frecuencia_Quijote'

valores_df = metricas_df
# T-test para dos grupos
grupos = valores_df['GrupoQS0-QSx'].unique()
grupo1 = valores_df[valores_df['GrupoQS0-QSx'] == grupos[0]][variable]
grupo2 = valores_df[valores_df['GrupoQS0-QSx'] == grupos[1]][variable]
t_stat, p_value = stats.ttest_ind(grupo1, grupo2)
print('Variable = ', variable, '\n')
print(f"Se utilizó el t-test de Student para muestras independientes.")
print(f"El estadístico t es {t_stat:.3f} y el valor p es {p_value:.4f}.")
if p_value < 0.05:
     print("Existe diferencia estadísticamente significativa entre los grupos de B.")
else:
    print("No se encontró diferencia estadísticamente significativa entre los grupos de B.")
print('\n******************************************')

#
#

#Frecuencia_Sancho
variable = 'Frecuencia_Sancho'

valores_df = metricas_df
# T-test para dos grupos
grupos = valores_df['GrupoQS0-QSx'].unique()
grupo1 = valores_df[valores_df['GrupoQS0-QSx'] == grupos[0]][variable]
grupo2 = valores_df[valores_df['GrupoQS0-QSx'] == grupos[1]][variable]
t_stat, p_value = stats.ttest_ind(grupo1, grupo2)
print('Variable = ', variable, '\n')
print(f"Se utilizó el t-test de Student para muestras independientes.")
print(f"El estadístico t es {t_stat:.3f} y el valor p es {p_value:.4f}.")
if p_value < 0.05:
     print("Existe diferencia estadísticamente significativa entre los grupos de B.")
else:
    print("No se encontró diferencia estadísticamente significativa entre los grupos de B.")
```
